## 3. Scaling Data 

```python
# Create scaler object using StandardScaler
scaler = StandardScaler()
```

```python
# Columns to exclude from scaling
columns_to_exclude = ['DATE', 'MONTH']
```

```python
# Select columns to scale
features_to_scale = [col for col in df.columns if col not in columns_to_exclude]
```

```python
# Create new dataframe with scaled data
df_scaled = pd.DataFrame(scaler.fit_transform(df[features_to_scale]), columns=features_to_scale, index=df.index)
```

```python
# Include the excluded columns in the scaled DataFrame
df_scaled = pd.concat([df[columns_to_exclude], df_scaled], axis=1)
```

## 4. Export Dataset

```python
# Export scaled dataframe
df_scaled.to_csv(os.path.join(path, 'Weather_Dataset_Scaled.csv'), index=False)
```
